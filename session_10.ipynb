```python
import numpy as np
import matplotlib.pyplot as plt
```

(I chose gamma distribution)

```python
shape_param = 2.0
scale_param = 2.0
```

```python
random_numbers = np.random.gamma(shape=shape_param, scale=scale_param, size=1000)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(random_numbers, bins=100, color='skyblue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers')
plt.savefig('gamma_distribution_histogram.pdf')
plt.show()
```
