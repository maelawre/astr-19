```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def compute_exp(x):
    return np.exp(x)
```

```python
y = compute_exp(x)

plt.figure()
plt.plot(x, y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Exponential Growth of Awesomeness")
plt.grid(True)
plt.savefig("exp_plot.pdf")
plt.show()
```
